```python
import os
import glob
```

```python
mat_files = glob.glob('tumor_data/mat_files/*')
```

```python
import h5py
from PIL import Image
import numpy as np

path_to_jpeg_files = 'tumor_data/jpeg_files'

for file_path in mat_files:
    with h5py.File(file_path) as fh:
        img = np.array(fh['cjdata']['image'])
        label = int(np.array((fh['cjdata']['label']))[0, 0]) - 1
        
    mat_file_name = file_path.split('/')[-1].split('.')[0]
    
    img = img * 255.0 / img.max()
    img = Image.fromarray(img)
    img = img.convert('RGB')
    
    filename = '{}_{}.jpeg'.format(mat_file_name, label)
    full_path = os.path.join(path_to_jpeg_files, filename)
        
    img.save(full_path)
```
