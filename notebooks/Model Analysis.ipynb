```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
with open('train_bninception_0_log.txt') as fh:
    bolai = fh.readlines()
```

```python
epochs = []
iters = []
losses = []
for bol in bolai:
    epoch, iteration, loss = tuple(bol.strip().split(' '))
    epochs.append(int(epoch))
    iters.append(int(iteration))
    losses.append(float(loss))
    
```

```python
plt.figure(figsize = (12,6))
plt.plot(range(1,len(iters)+1), losses)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('BNInception 1st Fold Training Losses Plot')
plt.legend(['Cross Entropy'])
```

```python
with open('test_bninception_0_log.txt') as fh:
    bolai = fh.readlines()

aepochs = []
acurs = []
alosses = []
for bol in bolai:
    epoch, iteration, loss = tuple(bol.strip().split(' '))
    aepochs.append(int(epoch))
    acurs.append(float(iteration))
    alosses.append(float(loss))
    
```

```python
plt.figure(figsize = (12,6))
plt.plot(aepochs, alosses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('BNInception 1st Fold Test Losses Plot')
plt.legend(['Cross Entropy'])
```

```python
errors = [100 - i for i in acurs]
plt.figure(figsize = (12,6))
plt.plot(aepochs, errors)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('BNInception 1st Fold Test Error Plot')
plt.legend(['Error (Percentage)'])
```
