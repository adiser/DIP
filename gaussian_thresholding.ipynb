```python
import cv2
import numpy as np
img = cv2.imread('data/sah/sah_fixed17.png')

frame = cv2.GaussianBlur(img, (9,9), 5)

kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpened_2 = cv2.filter2D(img, -1, kernel)

sharpened_1 = cv2.addWeighted(img, 1.5, frame, -0.5,  0)
```

```python
import matplotlib.pyplot as plt

plt.imshow(sharpened_1)
```
