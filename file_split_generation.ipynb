```python
from sklearn.model_selection import StratifiedKFold
import glob
import os
import numpy as np
```

```python
sah = glob.glob("data/sah/*")
sdh = glob.glob("data/sdh/*")

sah_targets = list(np.zeros(len(sah), dtype = int))
sdh_targets = list(np.ones(len(sdh), dtype = int))
```

```python
sah.extend(sdh)
sah_targets.extend(sdh_targets)
```

```python
X = sah
y = sah_targets
```

```python
skf = StratifiedKFold(n_splits = 5)
```

```python
for ii, (train, val) in enumerate(skf.split(X,y)):

    print('iteration', ii, train, val)
    
    train_file_name = "train_split_{}.txt".format(ii)
    
    with open(train_file_name, 'w+') as fh:
        for i in train:
            fh.write("{} {}\n".format(X[i], y[i]))
            
    val_file_name = "val_split_{}.txt".format(ii)
    
    with open(val_file_name, 'w+') as fh:
        for i in val:
            fh.write("{} {}\n".format(X[i], y[i]))
            
```
